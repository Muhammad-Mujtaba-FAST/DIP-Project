```python
from PIL import Image
import pytesseract
import cv2
import numpy as np

def extract_text_and_image(image_path):
    try:
        # Open the image file
        image = Image.open(image_path)

        # Use Tesseract OCR to extract text
        extracted_text = pytesseract.image_to_string(image)

        if extracted_text:
            # Display extracted text
            print("Extracted Text:")
            print(extracted_text)

            # Display the ID card image
            image.show()

            return extracted_text
        else:
            print("Text extraction failed. Check the image and Tesseract installation.")

    except Exception as e:
        print(f"Error: {str(e)}")

if __name__ == "__main__":
    # Provide the path to the image of the ID card
    image_path = "6.jpeg"

    # Call the function to extract text and display the image
    extracted_data = extract_text_and_image(image_path)

    if extracted_data:
        # Output the extracted data to a text file
        with open("output.txt", "w", encoding="utf-8") as file:
            file.write(extracted_data) 
```

```python
from PIL import Image, ImageDraw
import pytesseract

def extract_text_and_highlight(image_path):
    # Open the image file
    img = Image.open(image_path)

    # Use Tesseract to get bounding boxes of the recognized text
    boxes = pytesseract.image_to_boxes(img)

    # Draw rectangles around the text on the image
    draw = ImageDraw.Draw(img)
    for b in boxes.splitlines():
        b = b.split()
        x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
        draw.rectangle([x, img.size[1] - h, w, img.size[1] - y], outline="red", width=2)

    # Display the image with highlighted text
    img.show()

    # Use Tesseract to do OCR on the image
    extracted_text = pytesseract.image_to_string(img)

    # Print the extracted text
    print("Extracted Text:")
    print(extracted_text)

    # Save the extracted text to a file
    output_file_path = "extracted_text.txt"
    with open(output_file_path, 'w') as file:
        file.write(extracted_text)
    
    print(f"Extracted text saved to {output_file_path}")

if __name__ == "__main__":
    # Replace 'path/to/your/image.jpg' with the actual path to your image file
    image_path = '5.jpg'
    extract_text_and_highlight(image_path)
```

```python
from PIL import Image, ImageEnhance, ImageDraw
import pytesseract

def preprocess_image(image_path):
    img = Image.open(image_path)

    # Increase contrast
    enhancer = ImageEnhance.Contrast(img)
    img = enhancer.enhance(2.0)

    # Convert to grayscale
    img = img.convert('L')

    return img

def extract_text_and_highlight(image_path):
    # Preprocess the image
    img = preprocess_image(image_path)

    # Use Tesseract to get bounding boxes of the recognized text
    boxes = pytesseract.image_to_boxes(img, config='--psm 6', lang='eng')

    # Draw rectangles around the text on the image
    draw = ImageDraw.Draw(img)
    for b in boxes.splitlines():
        b = b.split()
        x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
        draw.rectangle([x, img.size[1] - h, w, img.size[1] - y], outline="red", width=2)

    # Display the image with highlighted text
    img.show()

    # Use Tesseract to do OCR on the preprocessed image
    extracted_text = pytesseract.image_to_string(img, config='--psm 6', lang='eng')

    # Print the extracted text
    print("Extracted Text:")
    print(extracted_text)

    # Save the extracted text to a file
    output_file_path = "extracted_text.txt"
    with open(output_file_path, 'w') as file:
        file.write(extracted_text)
    
    print(f"Extracted text saved to {output_file_path}")

if __name__ == "__main__":
    
    image_path = '5.jpg'
    extract_text_and_highlight(image_path)
```

```python
path = r'u2.jpeg'

img = cv2.imread(path, cv2.IMREAD_COLOR)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray,(3,3),0.0)


threshold, bw1 = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV) 
SE =  np.ones((2,15), dtype=np.uint8)
bw =  cv2.dilate(bw1,SE)
contours, hierarchy = cv2.findContours(bw,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
# filter contours based on height and width
temp = np.zeros_like(img)


## 7	Treat the image as a single text line
config = r'--oem 3 --psm 7'


for i in range(len(contours)):
    x,y,w,h = cv2.boundingRect(contours[i])
    aspectRatio = w/h
    # geometrical constraint
    if(aspectRatio>=2.0   and h>=10):
        cv2.rectangle(img, (x,y),(x+w,y+h),(0,255,0),2)
        crop = img[y:y+h, x:x+w]
        text = ocr.image_to_string(crop,config=config, lang='urd')
        print(text)

fig, axs = plt.subplots(1,2, figsize=(16,8))

axs[0].imshow(img,cmap='gray')
axs[0].axis('off')
axs[0].set_title('Original Image',fontsize='medium')

axs[1].imshow(bw,cmap='gray')
axs[1].axis('off')
axs[1].set_title('Original Image',fontsize='medium')

# print(text)
```
